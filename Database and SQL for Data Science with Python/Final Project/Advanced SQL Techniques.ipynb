# Exercise 1: Using Joins
You have been asked to produce some reports about the communities and crimes in the Chicago area. You will need to use SQL join queries to access the data stored across multiple tables.

### Question 1

##### Write and execute a SQL query to list the school names, community names and average attendance for communities with a hardship index of 98.

````python
%%sql
select CD.HARDSHIP_INDEX, CD.COMMUNITY_AREA_NAME, CPS.NAME_OF_SCHOOL, CPS.AVERAGE_STUDENT_ATTENDANCE
from CENSUS_DATA as CD
left outer join CHICAGO_PUBLIC_SCHOOLS as CPS
on CD.COMMUNITY_AREA_NUMBER = CPS.COMMUNITY_AREA_NUMBER
where CD.HARDSHIP_INDEX = 98
````

<details><summary>Click here for hints</summary>
1. Use tables CHICAGO_PUBLIC_SCHOOLS and CENSUS_DATA<br>
2. Use a left join<br>
3. can't use community area name for matching (case different)
</details>

### Question 2

##### Write and execute a SQL query to list all crimes that took place at a school. Include case number, crime type and community name.

````python
%%sql
select CSD. LOCATION_DESCRIPTION,CSD.CASE_NUMBER, CSD.PRIMARY_TYPE, CD.COMMUNITY_AREA_NAME
from CHICAGO_CRIME_DATA as CSD 
left outer join CENSUS_DATA as CD
on CD.COMMUNITY_AREA_NUMBER	 = CSD.COMMUNITY_AREA_NUMBER
where CSD.LOCATION_DESCRIPTION LIKE '%SCHOOL%'
````

(IN in_School_ID INTEGER, IN in_Leader_Score INTEGER)

BEGIN

 		UPDATE CHICAGO_PUBLIC_SCHOOLS

     			SET "Leaders_Icon" =  'average'
